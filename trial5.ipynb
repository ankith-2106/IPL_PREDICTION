```python
import pandas as pd

# Load the CSV file


# Initialize an empty dictionary to hold the results for each match and inning
match_data = {}

# Iterate over each match
for match_id in df['match_id'].unique():
    match_df = df[df['match_id'] == match_id]
    
    # Initialize a dictionary to store data for each inning in this match
    innings_data = {}
    
    # Iterate over each inning in the match
    for inning in match_df['inning'].unique():
        inning_df = match_df[match_df['inning'] == inning]
        
        # Initialize lists to store data for this inning
        over_list = []
        runs_scored_list = []
        wickets_list = []
        cumulative_score_list = []
        
        # Initialize cumulative score
        cumulative_score = 0
        
        # Iterate over each over in the inning
        for over in inning_df['over'].unique():
            over_df = inning_df[inning_df['over'] == over]
            
            # Calculate runs scored in this over
            runs_scored = over_df['total_runs'].sum()
            
            # Calculate the number of wickets in this over
            wickets = over_df['is_wicket'].sum()
            
            # Update cumulative score
            cumulative_score += runs_scored
            
            # Append the data to the lists
            over_list.append(over)
            runs_scored_list.append(runs_scored)
            wickets_list.append(wickets)
            cumulative_score_list.append(cumulative_score)
        
        # Store the lists in the innings_data dictionary
        innings_data[inning] = {
            'over': over_list,
            'runs_scored': runs_scored_list,
            'wickets': wickets_list,
            'cumulative_score': cumulative_score_list
        }
    
    # Store the innings data in the match_data dictionary
    match_data[match_id] = innings_data

# Display the results for a specific match_id and inning for demonstration
match_data[335982]  # Change the match_id to view other matches
```
